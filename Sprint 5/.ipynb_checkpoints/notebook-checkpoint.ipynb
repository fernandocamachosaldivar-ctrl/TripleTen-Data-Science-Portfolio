## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
megaline_calls['month'] = megaline_calls['call_date'].dt.month
calls_add = megaline_calls.groupby(['user_id', 'month']).agg(calls_made=('id', 'count')).reset_index()
print(calls_add.head())
```

```python
megaline_calls['month'] = megaline_calls['call_date'].dt.month
calls_add = megaline_calls.groupby(['user_id', 'month']).agg(calls_made=('id', 'count'),
                                                            minutes_used=('duration', 'sum')).reset_index()
print(calls_add.head())
```

```python
megaline_messages['month'] = megaline_messages['message_date'].dt.month# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.
messages_add = megaline_messages.groupby(['user_id', 'month']).agg(messages_sent=('id', 'count')
).reset_index()

print(messages_add.head())
```

```python
megaline_internet['month'] = megaline_internet['session_date'].dt.month
internet_add = megaline_internet.groupby(['user_id', 'month']).agg(mb_used_total=('mb_used', 'sum')
).reset_index()

print(internet_add.head())
```

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
monthly_consumption = pd.merge(calls_add, messages_add,on=['user_id', 'month'], how='outer')
monthly_consumption = pd.merge(monthly_consumption, internet_add, on=['user_id', 'month'], how='outer')
monthly_consumption = monthly_consumption.fillna(0)

print(monthly_consumption.head())
```

```python
user_plan_info = megaline_users[['user_id', 'plan', 'city']]
monthly_consumption = pd.merge(monthly_consumption, user_plan_info, on='user_id', how='left')
full_data = pd.merge(monthly_consumption, megaline_plans,left_on='plan', right_on='plan_name',
                     how='left')
full_data = full_data.drop('plan_name', axis=1)
print(full_data.head())
```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
def calculate_invoice (row):
    minutes_extra = 0
    minutes_dif = row['minutes_used'] - row['minutes_included']
    if minutes_dif > 0:
        minutes_extra = minutes_dif * row['usd_per_minute']
    messages_extra = 0
    messages_dif = row['messages_sent'] - row['messages_included']
    if messages_dif > 0:
        messages_extra = messages_dif * row['usd_per_message']
    data_extra = 0
    gb_used = row['mb_used_total'] / 1024
    gb_dif = gb_used - row['gb_per_month_included']
    if gb_dif > 0:
        gb_overage_billable = np.ceil(gb_dif)
        data_extra = gb_overage_billable * row['usd_per_gb']
    total_invoice = row['usd_monthly_pay'] + minutes_extra + messages_extra + data_extra 
    return total_invoice

full_data['monthly_revenue'] = full_data.apply(calculate_invoice, axis=1)
print(full_data[['user_id','month', 'minutes_used', 'messages_sent', 'mb_used_total', 'plan', 'monthly_revenue']].head())   
        
```

## Ingreso

[Del mismo modo que has estudiado el comportamiento de los usuarios, describe estadísticamente los ingresos de los planes.]

```python
def calculate_minutes_revenue(row):
    minutes_diff = row['minutes_used'] - row['minutes_included']
    if minutes_diff > 0:
        return minutes_diff * row['usd_per_minute']
    else:
        return 0
full_data['minutes_revenue'] = full_data.apply(calculate_minutes_revenue, axis=1)
minutes_revenue_stats = full_data.groupby('plan')['minutes_revenue'].agg(['mean', 'var', 
                                                                          'std', 'sum'])
print(minutes_revenue_stats)

```

```python

def calculate_messages_overage_revenue(row):
    messages_diff = row['messages_sent'] - row['messages_included']
    if messages_diff > 0:
        return messages_diff * row['usd_per_message']
    else:
        return 0

full_data['messages_overage_revenue'] = full_data.apply(calculate_messages_overage_revenue, axis=1)
message_revenue_stats = full_data.groupby('plan')['messages_overage_revenue'].agg(['mean', 'median', 'std', 'sum'])
print(message_revenue_stats)
```

```python
def calculate_data_revenue(row):
    gb_used = row['mb_used_total'] / 1024
    gb_diff = gb_used - row['gb_per_month_included']
    if gb_diff > 0: return np.ceil(gb_diff) * row['usd_per_gb']
    return 0
full_data['data_overage_revenue'] = full_data.apply(calculate_data_revenue, axis=1)
data_revenue_stats = full_data.groupby('plan')['data_overage_revenue'].agg(['mean', 'median', 'std', 'sum'])
print(data_revenue_stats)
```

```python
full_data['total_monthly_bill'] = (full_data['usd_monthly_pay'] +
                                   full_data['minutes_revenue'] +
                                   full_data['messages_overage_revenue'] +
                                   full_data['data_overage_revenue'])

monthly_avg_revenue = full_data.pivot_table(index='month',columns='plan',
                                            values='total_monthly_bill',aggfunc='mean')
monthly_avg_revenue.plot(kind='bar',title ='Ingreso Promedio Mensual por Usuario para Cada Plan',
                         xlabel='Mes', ylabel='Ingreso Promedio Mensual')
plt.legend(title='Plan')
plt.show()
```

Se puede observar que la principal fuente de ingresos son los usuarios que tienen el plan ultimate, sin embargo se muestra que los que tienen el plan surf tambien generan una gran cantidad de ingreso por los recargos al exceder su plan en cualquiera de los servicios de llamadas, mensajes o datos, el comportamiento de los ingresos del plan ultimate es constante debido a que porr lo general no tienen cargos extra, en cambio los usuarios que tienen el plan Surf es variable y dependiente de si el usuario tiene que utilizar mas de lo que le proporciona su plan

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
surf_revenue = full_data[full_data['plan'] == 'surf']['total_monthly_bill']
ultimate_revenue = full_data[full_data['plan'] == 'ultimate']['total_monthly_bill']
alpha = 0.05
results = st.ttest_ind(surf_revenue, ultimate_revenue, equal_var=False)
print('valor p: ', results.pvalue)
if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula: El ingreso promedio entre los planes es diferente.")
else:
    print("No podemos rechazar la hipótesis nula: No hay evidencia suficiente para decir que los ingresos promedio son diferentes.")
```
